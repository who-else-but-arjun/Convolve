```python
def replace_rare_values(cdna_df, column, threshold=20):
    value_counts = cdna_df[column].value_counts()
    rare_values = value_counts[value_counts < threshold].index
    cdna_df[column] = cdna_df[column].replace(rare_values, 'Others')
    return cdna_df

cdna_df = replace_rare_values(cdna_df, 'v29', threshold=20)
cdna_df = replace_rare_values(cdna_df, 'v55', threshold=21)
```

```python
object_columns = cdna_df.select_dtypes(include=['object']).columns
for col in object_columns:
    mode = cdna_df[col].mode()[0]  
    cdna_df[col].fillna(mode, inplace=True)
numeric_columns = cdna_df.select_dtypes(include=['number']).columns
cdna_df[numeric_columns] = cdna_df[numeric_columns].interpolate(method='linear', axis=0,limit_direction='both')
```

```python
int_columns = cdna_df.select_dtypes(include=['int64', 'int32']).columns
scaler = MinMaxScaler()
cdna_df[int_columns] = scaler.fit_transform(cdna_df[int_columns])

float_columns = cdna_df.select_dtypes(include=['float64']).columns
scaler = StandardScaler()
cdna_df[float_columns] = scaler.fit_transform(cdna_df[float_columns])
```

```python
cdna_df['v60'] = cdna_df['v60'].apply(lambda x: 1 if x == 'N' else 0)
cdna_df['v60'] = cdna_df['v60'].astype(int)
boolean_columns = cdna_df.select_dtypes(include=['bool']).columns
for col in boolean_columns:
    cdna_df[col] = cdna_df[col].astype(int)
exclude_columns = ['CUSTOMER_CODE']
object_columns = [col for col in cdna_df.select_dtypes(include=['object']).columns if col not in exclude_columns]
label_encoders = {}
for col in object_columns:
    le = LabelEncoder()
    cdna_df[col] = le.fit_transform(cdna_df[col])
    label_encoders[col] = le  
```

```python
time_slots = [
    (9, 12), (12, 15), (15, 18), (18, 21)  
]

def get_time_slot(timestamp):
    if pd.isnull(timestamp):
        return None  
    try:
        dt = datetime.strptime(timestamp, '%Y-%m-%dT%H:%M:%S.%fZ')
        weekday = dt.weekday()  
        hour = dt.hour
        for i, (start, end) in enumerate(time_slots):
            if start <= hour < end:
                day_slot = weekday * 4 + (i + 1) 
                return day_slot
        return None
    except Exception:
        return None

hist_df = hist_df.dropna(subset=['send_timestamp'])
hist_df['send_timestamp'] = hist_df['send_timestamp'].apply(get_time_slot)
hist_df['send_timestamp'].fillna(hist_df['send_timestamp'].mode()[0], inplace=True)
hist_df['send_timestamp'] = hist_df['send_timestamp'].astype('int64')
hist_df['open_timestamp'] = hist_df['open_timestamp'].apply(get_time_slot)
hist_df['open_timestamp'].fillna(0, inplace=True)
hist_df['open_timestamp'] = hist_df['open_timestamp'].astype('int64')
hist_df = hist_df.rename(columns={'customer_code': 'CUSTOMER_CODE'})
```
